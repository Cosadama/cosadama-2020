### Î∂àÏö©Ïñ¥ Stopwords  
Ïú†ÏùòÎØ∏Ìïú Îã®Ïñ¥ ÌÜ†ÌÅ∞ÎßåÏùÑ ÏÑ†Î≥ÑÌïòÍ∏∞ ÏúÑÌï¥ÏÑú ÌÅ∞ ÏùòÎØ∏Í∞Ä ÏóÜÎäî Îã®Ïñ¥ ÌÜ†ÌÅ¥ÏùÑ Ï†úÍ±∞ÌïòÎäî ÏûëÏóÖÏûÖÎãàÎã§. ÏòàÎ•º Îì§Ïñ¥, I, my, me, over, Ï°∞ÏÇ¨, Ï†ëÎØ∏ÏÇ¨ Í∞ôÏùÄ Îã®Ïñ¥Îì§ÏùÄ ÏûêÏ£º Îì±Ïû•ÌïòÎÇò Ïã§Ï†ú ÏùòÎØ∏ Î∂ÑÏÑùÏóê Í∏∞Ïó¨ÌïòÎäî Î∞îÍ∞Ä Í±∞Ïùò ÏóÜÍ≥†, Ïù¥Î•º Î∂àÏö©Ïñ¥ÎùºÍ≥† Î∂àÎü¨Ïöî.

[Îçî ÏïåÏïÑÎ≥¥Í≥† Ïã∂Îã§Î©¥ Ïó¨Í∏∞Î°ú](https://medium.com/@omicro03/%EC%9E%90%EC%97%B0%EC%96%B4%EC%B2%98%EB%A6%AC-nlp-5%EC%9D%BC%EC%B0%A8-%EB%B6%88%EC%9A%A9%EC%96%B4-277694095a6f)

ÌõÑÏóêÎèÑ Î∂àÏö©Ïñ¥ ÏûëÏóÖÏùÑ Í≥ÑÏÜçÌïòÍ≤†ÏßÄÎßå Ïù¥Í≤å ÏÉùÍ∞ÅÎ≥¥Îã§ Îß§Ïö∞ Í∑ÄÏ∞ÆÏùÄ ÏûëÏóÖÎãàÎã§ „Öé„Öé„Öé„Öé ÏòÅÏñ¥ Î∂àÏö©Ïñ¥Îäî Î™®Îìà ÏûêÏ≤¥Ïóê ÎÇ¥Ïû•ÎêòÏñ¥ ÏûàÏßÄÎßå, ÌïúÍ∑πÏùÄ ÏßÅÏ†ë Î∂àÏö©Ïñ¥Î•º ÏûÖÎ†•Ìï¥Ï§òÏïº ÌïòÍ±∞Îì†Ïöî... „Öé„Öé„Öé„Öé


```python
stopwords = set(STOPWORDS)

# Ïï®Î¶¨Ïä§ ÏÜåÏÑ§Ïóê saidÍ∞Ä ÎßéÏù¥ Îì±Ïû•Ìï¥ÏÑú Î∂àÏö©Ïñ¥ Ï≤òÎ¶¨
stopwords.add("said")
```

ÏõåÎìú ÌÅ¥ÎùºÏö∞Îìú ÏûêÏ≤¥ÏóêÏÑú Îã®Ïñ¥ ÌöüÏàòÎ•º ÏÑ∏Îäî Í∏∞Îä•ÏùÑ Ï†úÍ≥µÌï©ÎãàÎã§. Ïù¥Î•º Ïù¥Ïö©Ìï¥ÏÑú ÏµúÎπà Îã®Ïñ¥Î•º Ï∞æÏùÑÍ±∞ÏóêÏöî

```python
wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask,
              stopwords = stopwords)
wc = wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
# imshow ÌïòÎäî Îç∞Ïù¥ÌÑ∞Îäî alice_mask Î≥ÄÏàò (Ïï®Î¶¨Ïä§ Ïù¥ÎØ∏ÏßÄÎ•º Îã¥ÏïÑÏ§Ä Î≥ÄÏàò)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
# imshowÌïòÎäî Í≤ÉÏùÄ wc (wordcloudÎ™®ÎìàÏùÑ ÌôúÏö©Ìï¥ ÏÉùÏÑ±Ìïú Í∞ùÏ≤¥)
plt.axis('off')
plt.show()
```

### Ïä§ÌÉÄÏõåÏ¶àÏùò A New Hope ÌÖçÏä§Ìä∏ ÏãúÍ∞ÅÌôî

```python
text = open('09. a_new_hope.txt').read()

# ÏôºÏ™Ω Îã®Ïñ¥Î•º Ïò§Î•∏Ï™ΩÏúºÎ°ú Î≥ÄÍ≤Ω 
text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", 'Luke')

mask = np.array(Image.open('09. stormtrooper_mask.png'))
```

```python
# Î∂àÏö©Ïñ¥ Ï≤òÎ¶¨ 
stopwords = set(STOPWORDS)
stopwords.add('int')
stopwords.add('ext')
```

```python
wc = WordCloud(max_words=1000, mask=mask, stopwords=stopwords,
               margin=10, random_state=1).generate(text)
default_colors = wc.to_array()
```

```python
# ÌÖçÏä§Ìä∏ ÏÉâ ÌöåÏÉâÏúºÎ°ú Ï±ÑÏö∞Îäî Ìï®Ïàò
import random
def grey_color_func(word, font_size, position, orientation, 
                    random_state=None, **kwargs): 
                    return 'hsl(0, 0%%, %d%%)' % random.randint(60,100)
```

**ÏûêÎ£åÎ•º Í∞ÄÏ†∏Ïò§Í≥† -> Î∂àÏö©Ïñ¥Î•º Ï≤òÎ¶¨ -> ÏõåÎìúÌÅ¥ÎùºÏö∞Îìú ÏïàÏóê Îã¥ÏïÑÏ£ºÍ≥† -> ÏãúÍ∞ÅÌôî** Ïù¥Îü∞ ÏãùÏúºÎ°ú ÌùòÎü¨Í∞ÄÍ≥† ÏûàÏäµÎãàÎã§

### Ïú°ÏïÑÌú¥ÏßÅ Í¥ÄÎ†® Î≤ïÏïàÏóê ÎåÄÌïú Î∂ÑÏÑù  
KoNLPyÍ∞Ä ÎÇ¥Ïû•ÌïòÍ≥† ÏûàÎäî Î≤ïÎ•† Î¨∏ÏÑú Ï§ë ÌïòÎÇòÎ•º Í∞ÄÏßÄÍ≥† Í∏∞Ï¥àÏ†ÅÏù∏ Î∂ÑÏÑùÏùÑ Ìï¥Î≥¥Îäî ÌäúÌÜ†Î¶¨ÏñºÏù¥ÏóêÏöî. Îß§Ïö∞ Í∞ÑÎã®Ìï©ÎãàÎã§.

```python
import nltk
from konlpy.corpus import kobill
files_ko = kobill.fileids()
# konlpyÏùò ÎÇ¥Î∂ÄÎ¨∏ÏÑú Ï§ë Ïú°ÏïÑÌú¥ÏßÅ Í¥ÄÎ†® Î≤ïÏïà Ï†ú 1809890Ìò∏ ÏùΩÍ∏∞
doc_ko = kobill.open('1809890.txt').read()
```

```python
doc_ko
```

ÏïÑÍπå Î∞∞Ïõ†Îçò okt Î™®ÎìàÎ°ú Î™ÖÏÇ¨Î•º Î∂ÑÏÑùÌï¥Î¥ÖÏãúÎã§!

```python
from konlpy.tag import Okt 
okt = Okt()
# doc_koÏùò ÎÇ¥Ïö©ÏùÑ oktÎ™®ÎìàÎ°ú Î™ÖÏÇ¨Î∂ÑÏÑùÏùÑ ÌïòÍ≥†, Í∑∏ Í≤∞Í≥ºÎ•º tokens_koÎùºÎäî Í∞ùÏ≤¥Ïóê Îã¥ÏïÑÏ§ÄÎã§
tokens_ko = okt.nouns(doc_ko)
tokens_ko
```

- Î¨∏ÏûêÏó¥ Îã®ÏúÑÎ•º ÌÜ†ÌÅ∞(token)Ïù¥ÎùºÍ≥† ÌïòÍ≥† Ïù¥Î†áÍ≤å Î¨∏ÏûêÏó¥ÏùÑ ÌÜ†ÌÅ∞ÏúºÎ°ú ÎÇòÎàÑÎäî ÏûëÏóÖÏùÑ ÌÜ†ÌÅ∞ ÏÉùÏÑ±(tokenizing)Ïù¥ÎùºÍ≥† ÌïòÍ≥†
- Î¨∏ÏûêÏó¥ÏùÑ ÌÜ†ÌÅ∞ÏúºÎ°ú Î∂ÑÎ¶¨ÌïòÎäî Ìï®ÏàòÎ•º ÌÜ†ÌÅ∞ ÏÉùÏÑ± Ìï®Ïàò(tokenizer)ÎùºÍ≥† Ìï©ÎãàÎã§. 

Ïù¥ ÎÇ¥Ïö©ÏùÑ Î®∏Î¶¨Ïóê Îã¥ÏïÑÎëêÍ≥† ÏΩîÎìúÎ•º ÏùΩÏñ¥Î≥¥Î©¥ ÏΩîÎìúÏùò ÌùêÎ¶ÑÏù¥ Îçî Ïûò ÏùΩÌûêÍ≤ÅÎãàÎã§

```python
ko = nltk.Text(tokens_ko, name='ÎåÄÌïúÎØºÍµ≠ Íµ≠Ìöå ÏùòÏïà Ï†ú 1809890Ìò∏')
print(len(ko.tokens)) # ÏàòÏßëÎêú ÌÜ†ÌÅ∞(Îã®Ïñ¥)Ïùò ÌöüÏàò
print(len(set(ko.tokens))) # ÏàòÏßëÎêú Îã®Ïñ¥ Ï§ë Í≥†Ïú†Ìïú(unique) ÌÜ†ÌÅ∞Ïùò ÌöüÏàò
ko.vocab()
```

```python
plt.figure(figsize=(12,6))
ko.plot(50) # ÏÉÅÏúÑ 50 ÌÜ†ÌÅ∞Î•º Ï†ïÎ†¨ÌïòÏó¨ Î≥¥Ïó¨Ï§å 
plt.show()
```

Ïú°ÏïÑÌú¥ÏßÅÏù¥ÎùºÎäî Îã®Ïñ¥Í∞Ä ÎßéÏù¥ Îì±Ïû•ÌïòÍ≥†, Îßå, Ïï°, ÏÑ∏ Îì±Í≥º Í∞ôÏùÄ Î∂àÏö©Ïñ¥ÎèÑ ÎßéÏù¥ Î≥¥ÏûÖÎãàÎã§. Í∑∏ÎûòÏÑú ÌïòÎÇòÏî© Î∂àÏö©Ïñ¥Î•º ÏßÅÏ†ë Ï†úÍ±∞Ìï¥Ï§òÏïºÍ≤†Ï£†? Îã®Ïñ¥ Î¶¨Ïä§Ìä∏Î•º Î≥¥Í≥† Î≥∏Ïù∏Ïù¥ ÏßÅÏ†ë Î∂àÏö©Ïñ¥ÎùºÍ≥† ÏÉùÍ∞ÅÌïòÎäî Í≤ÉÎì§ÏùÑ ÎÑ£Ïñ¥ÏÑú Ï†úÍ±∞Ìï¥Ï£ºÎ©¥ Îê©ÎãàÎã§. wordcloudÎ•º ÏãúÍ∞ÅÌôîÌïòÎäî Í≤ÉÏùÄ Ïù¥Î†áÍ≤å ÌíàÏù¥ ÎßéÏù¥ Îì§Ïñ¥Í∞ÄÎäî ÏùºÏù¥ÏóêÏöî...„Öú„Öú

```python
stop_words = ['.','(',')',',',"'",'%','-','X',').',
              'x','Ïùò','Ïûê','Ïóê','Ïïà','Î≤à','Ìò∏','ÏùÑ','Ïù¥','Îã§','Îßå','Î°ú','Í∞Ä','Î•º']
ko = [each_word for each_word in ko if each_word not in stop_words]
ko
```

```python
ko = nltk.Text(ko, name='ÎåÄÌïúÎØºÍµ≠ Íµ≠Ìöå ÏùòÏïà Ï†ú 1809890Ìò∏')
# Î∂àÏö©Ïñ¥Î•º Ï≤òÎ¶¨Ìïú Ïù¥ÌõÑ Îã§Ïãú ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```
