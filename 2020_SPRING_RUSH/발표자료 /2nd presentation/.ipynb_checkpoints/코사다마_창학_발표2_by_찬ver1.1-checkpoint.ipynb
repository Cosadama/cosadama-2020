### 절대 주소와 상대 주소
 - 절대 주소: 그냥 누르면 이동되는 주소.
   - e.g.) http://naver.com
 - 상대 주소: 부분을 나누어 표현한 주소. a와 b처럼 두개의 부분으로 나눌 수도 있고, 상황에 따라 n개로 나눌 수도 있겠다.
   - e.g.) 
     - a: http://comic.     
     - b: naver.com    

**상대 주소는 왜 나눌까?**

#### urllib의 urljoin: 상대경로 앞에만 특정 주소를 붙여 절대경로로 전환

```python
from urllib.parse import urljoin

urljoin('https://www.chicagomag.com', item.find('a')['href'])
```

```python
rank=[]
main_menu =[]
cafe_name=[]
url_add=[]

list_soup = soup.find_all('div', 'sammy')

for item in list_soup:
    rank.append(item.find(class_='sammyRank').get_text())
    
    tmp_string = item.find(class_='sammyListing').get_text()
    
    main_menu.append(re.split(('\r\n|\n'),tmp_string)[0])
    cafe_name.append(re.split(('\r\n|\n'),tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

#### dataframe에 정보 담기

```python
import pandas as pd

data = {'Rank':rank, 'Menu':main_menu, 'Cafe':cafe_name, 'URL':url_add}
df = pd.DataFrame(data)
df.head()
```

```python
df = pd.DataFrame(data, columns=['Rank', 'Cafe', 'Menu', 'URL'])
df.head()
```

```python
price = []
address = []

for n in df.index[:3]:
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'html.parser')

    gettings = soup_tmp.find('p', 'addy').get_text()
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
```

#### tqdm 모듈을 사용해 50페이지에 접근하기

```python
from tqdm import tqdm_notebook
```

```python
price = []
address =[]

for n in tqdm_notebook(df.index):
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'html.parser')
    
    gettings = soup_tmp.find('p', 'addy').get_text()
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
    
```

#### google maps와 Folium으로 맛집 지도 표식하기

```python
import pandas as pd
import numpy as np
import googlemaps
import folium
from tqdm import tqdm_notebook
```

```python
df = pd.read_csv('04. best_sandwitches_list_chicago2.csv', index_col=0)
df.head()
```

```python
gmaps_key = '******************************'
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
target_name = df['Address'][1] + ', '+ 'Chicago'
gmaps_output = gmaps.geocode(target_name)
gmaps_output
```

```python
gmaps_output[0]
```

```python
gmaps_output[0].get('geometry')
```

```python
gmaps_output[0].get('geometry')['location']
```

```python
gmaps_output[0].get('geometry')['location']['lng']
```

```python
lat =[]
lng =[]

for n in tqdm_notebook(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n] + ', '+ 'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
lat[:15], lng[:15]
```

```python
df['lat'] = lat
df['lng'] = lng
df[['lat','lng']].head()
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],
                    zoom_start=11)
folium.Marker([df['lat'].mean(), df['lng'].mean()],
             popup='center').add_to(mapping)

mapping
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],
                    zoom_start=11)
for n in df.index:
    if df['Address'][n] !='Multiple':
        folium.Marker([df['lat'][n], df['lng'][n]],
                     popup=df['Cafe'][n]).add_to(mapping)
        
mapping
```
