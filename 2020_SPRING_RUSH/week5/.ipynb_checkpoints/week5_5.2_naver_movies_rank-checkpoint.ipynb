### 100 í˜ì´ì§€ í¬ë¡¤ë§ í•˜ê¸°

ê·¸ëŸ¼ í•´ë‹¹ ì •ë³´ë“¤ì´ ëª¨ë‘ ì–´ë””ì— ë‹´ê²¨ì ¸ ìˆëŠ”ì§€ ì•Œì•˜ìœ¼ë‹ˆ, ì´ë¥¼ ì´ì œ ê°€ì ¸ì™€ì•¼ í•˜ê² ì£ . ê·¸ ì „ì— urllib.parse.quoteì— ëŒ€í•´ ì•Œì•„ë³¼ ê²ƒì¸ë°ìš”.

```python
import urllib 

url = urllib.parse.quote('íŒŒì´ì¬')
url
```

ì´ë ‡ê²Œ urllib.parse.quoteëŠ” í…ìŠ¤íŠ¸(í•œê¸€ì´ë‚˜ ìˆ«ì)ë¥¼ urlì— ì í•©í•œ ìš©ì–´ë¡œ ë°”ê¿”ì£¼ëŠ” ê²ƒì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ì´ê²ƒì´ ì™œ í•„ìš”í•˜ëƒë©´, url ì£¼ì†Œì— ë‚ ì§œê°€ ì í˜€ìˆìœ¼ë¯€ë¡œ ë‚ ì§œë¥¼ ìš©ì´í•˜ê²Œ ë°”ê¿”ì£¼ë©´ì„œ í¬ë¡¤ë§ í•˜ê¸° ìœ„í•´ì„œì—ìš”. 

- https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20200327

ì§€ë‚œ ì‹œê°„ì— í–ˆë˜ ê²ƒì´ ê¸°ì–µë‚  ê²ƒ ê°™ì€ë°ìš”, ì—¬ê¸°ì— ë§¨ ë’¤ dateë¥¼ ë³€ê²½í•´ë³´ì„¸ìš”. ê·¸ëŸ¬ë©´ ê·¸ ì¼ìì— ë§ì¶°ì„œ ì˜í™” í‰ì  ìˆœì´ ë°”ë€Œê²Œ ë©ë‹ˆë‹¤. ì•ì„œ ë§Œë“¤ì–´ë‘” dateë¥¼ í™œìš©í•´ urlì„ ë³€ê²½í•´ì£¼ì–´ ì´ 100 í˜ì´ì§€ì—ì„œ í¬ë¡¤ë§ì„ í•´ì˜¬ ê²ƒì´ëë‹ˆë‹¤.

```python
import urllib
from tqdm import tqdm_notebook

movie_date = []
movie_name = []
movie_point = []

date = pd.date_range('2019-12-01', periods=100, freq='D')

for today in tqdm_notebook(date):
    html = "http://movie.naver.com/" + \
                                    "movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date=
                                   urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, "html.parser")
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
```

ê·¸ëƒ¥ dataë¥¼ quoteì— ë„£ì–´ì£¼ë©´ errorê°€ ëœ¨ëŠ”ë°ìš”, url ë‚ ì§œ ì–‘ì‹ì— ë§ë„ë¡ í˜•ì‹ì„ ì €ì¥í•´ì£¼ëŠ” ê²ƒì´ í•„ìš”í•©ë‹ˆë‹¤. í˜•ì‹ì— ë§ì¶°ì„œ ë…„-ì›”-ì¼ì´ ë‚˜íƒ€ë‚˜ë„ë¡ í•´ì£¼ëŠ” ê²ƒì´ .surftimeì´ì—ìš”.

ë˜í•œ ì—¬ê¸°ì„œëŠ” appendê°€ ì•„ë‹Œ extend()ë¥¼ ì‚¬ìš©í–ˆì£ . ì°¨ì´ì ì€ ì•„ë˜ ë§í¬ë¥¼ ë³´ì‹œë©´ ë©ë‹ˆë‹¤. 
- [extend()ì™€ append()ì˜ ì°¨ì´ì ](https://hashcode.co.kr/questions/23/íŒŒì´ì¬ì—-appendì™€-extendì˜-ì°¨ì´ì )

ê·¸ëŸ¼ ì˜ ê°€ì ¸ì™€ì¡ŒëŠ”ì§€ í™•ì¸í•´ë³´ê² ìŠµë‹ˆë‹¤.

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name,
                     'point':movie_point})
movie.head(10)
```

(êµì¬ëŠ” í•©ì‚°ì´ì§€ë§Œ ì €ëŠ” í‰ê· ì´ ë” ë‚«ê² ë‹¤ ìƒê°í–ˆì–´ìš”. í•©ì‚°ìœ¼ë¡œ í•  ê²½ìš° ì˜í™”ê°€ ê±¸ë¦° ì¼ ìˆ˜ì— ì˜í–¥ì„ ë§ì´ ë°›ìœ¼ë‹ˆê¹Œìš”)

### ì˜í™”ë³„ ë‚ ì§œ ë³€í™”ì— ë”°ë¥¸ í‰ì  ë³€í™” ì‹œê°í™” 

ì´ì œëŠ” ë‚ ì§œì— ë”°ë¼ì„œ ì˜í™”ì˜ í‰ì  ë³€í™”ê°€ ì–´ë–»ê²Œ ë‚˜íƒ€ë‚˜ëŠ”ì§€ í™•ì¸í•´ë³´ë ¤ê³  í•´ìš”. ì—¬ê¸°ì—ì„œë„ pivot_tableì„ ì‚¬ìš©í•˜ë ¤ í•©ë‹ˆë‹¤.
