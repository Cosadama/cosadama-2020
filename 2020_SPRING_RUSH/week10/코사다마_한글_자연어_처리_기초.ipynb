## Naive Bayes ClassifierÏùò Ïù¥Ìï¥ - ÏòÅÎ¨∏

Î®∏Ïã†Îü¨ÎãùÏóêÎäî ÏßÄÎèÑÌïôÏäµÍ≥º ÎπÑÏßÄÎèÑÌïôÏäµÏù¥ ÏûàÎäîÎç∞Ïöî, Ïù¥ ÎÇòÏù¥Î∏å Î≤†Ïù¥Ï¶à Î∂ÑÎ•òÍ∏∞Îäî ÏßÄÎèÑÌïôÏäµÏóê ÏÜçÌï©ÎãàÎã§. Ïª¥Ìì®ÌÑ∞Î•º ÌïôÏäµÏãúÌÇ¨ Îïå Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú Î†àÏù¥Î∏îÏù¥ ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê(Ï†ïÎãµÏù¥ ÏûàÎã§!) 'ÏßÄÎèÑ'(supervised)ÌïôÏäµÏù¥Îùº Î∂ÄÎ¶ÖÎãàÎã§. 

Î®∏Ïã†Îü¨ÎãùÏùÄ Í∏∞Î≥∏Ï†ÅÏúºÎ°ú Ïñ¥Îñ†Ìïú ÏïåÍ≥†Î¶¨Ï¶òÏùÑ ÎßåÎì§Í∏∞ ÏúÑÌï¥ ÌïôÏäµÏãúÌÇ¨ train dataÏôÄ Ïù¥ train dataÎ•º ÌÜµÌï¥ ÌïôÏäµÌïú ÏïåÍ≥†Î¶¨Ï¶òÏúºÎ°ú ÏÉàÎ°≠Í≤å ÏòàÏ∏°Ìï¥Î≥º test dataÎ•º ÏöîÍµ¨Ìï©ÎãàÎã§. Ïó¨Í∏∞ÏóêÏÑúÎèÑ trainÏùÑ ÌÜµÌï¥ ÏïåÍ≥†Î¶¨Ï¶òÏùÑ ÌïôÏäµÏãúÌÇ§Í≥† Ïù¥Î•º ÏÉàÎ°úÏö¥ test dataÎ•º ÎÑ£Ïñ¥ ÏòàÏ∏°Ìï¥Î≥º Í≤ÅÎãàÎã§.

```
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```
# Ïó∞ÏäµÏö© Îç∞Ïù¥ÌÑ∞ train 
train = [('i like you','pos'),
         ('i hate you', 'neg'),
         ('you like me', 'neg'),
         ('i like her',' pos')]
```

trainÏóê ÏûàÎäî Î™®Îì† Îã®Ïñ¥Îì§ÏùÑ Í∞ÄÏ†∏Ïò§Îäî ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÍ∏∞Ïóê ÏïûÏÑú Ïù¥Ìï¥Î•º ÏúÑÌï¥ ÏïÑÎûò ÏΩîÎìúÎ•º Ï†úÏãúÌï¥Î¥§Ïñ¥Ïöî. SetÏúºÎ°ú Í∞êÏã∏Ï£ºÎäî Ïù¥Ïú†Îäî Ï§ëÎ≥µÏùÑ ÏóÜÏï†Í∏∞ ÏúÑÌï¥ÏÑú!

```
all_words = set(word.lower() for sentence in train 
                for word in word_tokenize(sentence[0]))
all_words
```

ÏïûÏùò ÎßêÎ≠âÏπò(all_words)Î•º Í∏∞Ï§ÄÏúºÎ°ú train Î¨∏Ïû•Ïóê ÏÜçÌïú Îã®Ïñ¥Ïù∏ÏßÄ ÏïÑÎãåÏßÄÎ•º Í∏∞Î°ùÌïòÍ∏∞

```
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1])
          for x in train]
t
```

Ï≤´Î≤àÏß∏ Î¨∏Ïû•ÏóêÎäî hateÎäî ÏóÜÍ≥†, iÎäî ÏûàÎã§Îäî Îúª! ÏΩîÎìúÍ∞Ä Ï°∞Í∏à Î≥µÏû°Ìï¥ÏÑú Ïù¥Ìï¥ÌïòÍ∏∞ Ïñ¥Î†§Ïö∏ Ïàò ÏûàÏßÄÎßå, ÌôúÏö©Ìï† Ïàò ÏûàÏúºÎ©¥ ÎêòÍ≤†Ï£µ„Öé„Öé Ïù¥Í≤ÉÏùÑ ÌôúÏö©Ìï¥ ÎÇòÏù¥Î∏å Î≤†Ïù¥Ï¶à Î∂ÑÎ•òÍ∏∞Î•º ÎèôÏûëÏãúÌÇ§Í≤†ÏäµÎãàÎã§.

```
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

ÌïúÍ∏ÄÎ°úÎèÑ ÌïòÎ†§Î©¥ ÌòïÌÉúÏÜå Î∂ÑÏÑùÏùÑ Í±∞Ï≥êÏïº Ìï¥Ïöî. ÌòïÌÉúÏÜå Î∂ÑÏÑùÏùÑ Í±∞ÏπòÏßÄ ÏïäÏùÑ Í≤ΩÏö∞ Ïñ¥ÎñªÍ≤å ÎêòÎäîÏßÄÎ∂ÄÌÑ∞ ÏÇ¥Ìé¥Î¥ÖÏãúÎã§.



```
from konlpy.tag import Okt
```

```
pos_tagger = Okt()
```

```
def tokenize(doc): 
  return ['/'.join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]
```

konlpyÎ•º ÎßåÎì† Lucy ParkÏóê Îî∞Î•¥Î©¥ ÏúÑÏùò Ìï®ÏàòÏ≤òÎüº ÌÉúÍ∑∏Î•º Î∂ôÏó¨Ï£ºÎäî Í≤ÉÏù¥ Ïú†Î¶¨ÌïòÎã§Í≥† Ìï¥Ïöî. tokenize Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥ÏÑú train Î¨∏Ïû•ÏùÑ ÌòïÌÉúÏÜå Î∂ÑÏÑùÌï¥Î≥¥Í≤†ÏäµÎãàÎã§.

```
train_docs = [(tokenize(row[0]), row[1]) for row in train]
train_docs
```

```
# Ïù¥ ÏΩîÎìúÎ•º ÏïÑÎûò ÏΩîÎìúÏóêÏÑúÎäî Ìïú Ï§ÑÎ°ú Ï†ÅÏñ¥Ï§Ä Í≤ÅÎãàÎã§!
for d in train_docs: 
  for t in d[0]: 
    print(t)
```

```
tokens = [t for d in train_docs for t in d[0]]
tokens
```

Ïù¥Î†áÍ≤å Ï†ÑÏ≤¥ ÎßêÎ≠âÏπòÎ•º ÎßåÎì§ÏóàÏäµÎãàÎã§. ÏïÑÎûò Ìï®ÏàòÎäî Ìï¥Îãπ Î¨∏Ïû•Ïóê ÏûàÎäî ÌòïÌÉúÏÜåÍ∞Ä tokens ÏïàÏóê Îì§Ïñ¥ÏûàÎäîÏßÄ ÌôïÏù∏ÌïòÎäî Í≤ÉÏù¥ÏóêÏöî.


```
def term_exists(doc): 
  return {word: (word in set(doc)) for word in tokens}
```

```
import scipy as sp 

def dist_raw(v1, v2):
  delta = v1 - v2 
  return sp.linalg.norm(delta.toarray())
```

```
best_doc = None
# 65535Í∞Ä Î¨¥Ïä® ÏùòÎØ∏Ïù∏ÏßÄÎäî Î™®Î•¥Í≤†ÎÑ§Ïöî... 
best_dist = 65535 
best_i = None 

for i in range(0, num_samples):
  post_vec = X.getrow(i)
  d = dist_raw(post_vec, new_post_vec)

  # iÎäî Î¨∏Ïû• Î≤àÌò∏, dÎäî Ïú†ÏÇ¨ÎèÑ
  print('== Post %i with dist=%.2f    : %s' % (i,d,contents[i]))

# best_distÍ∞Ä Í≥ÑÏÜç ÎÇÆÍ≤å ÏàòÏ†ïÎêòÎäî Í≤É! (ÎÇÆÏùÑÏàòÎ°ù Ïú†ÏÇ¨ÎèÑ ÎÜíÏùå)
  if d < best_dist: 
    best_dist = d
    best_i = i
```

```
def dist_norm(v1, v2): 
  v1_normalized = v1 / sp.linalg.norm(v1.toarray())
  v2_normalized = v2 / sp.linalg.norm(v2.toarray())

  delta = v1_normalized - v2_normalized 

  return sp.linalg.norm(delta.toarray())
```

```
best_doc = None 
best_dist = 65535
best_i = None 

for i in range(0, num_samples): 
  # ÏïûÏóêÏÑú arrayÎ°ú Î∞îÍøîÏ§¨ÏúºÎãà Îã§Ïãú arrayÎ°ú Î∞îÍøîÏ§ÑÌïÑÏöî ÏóÜÏùå 
  post_vec = X.getrow(i)
  d = dist_norm(post_vec, new_post_vec)

  print("== Post %i with dist=%2f   : %s" % (i,d,contents[i]))

  if d < best_dist: 
    best_dist = d
    best_i = i 
```

```
print('Best post is %i, dist = %.2f' % (best_i, best_dist))
print('-->', new_post)
print('---->', contents[best_i])
```
