ì´ì œ ì¡°ê¸ˆ ì´í•´ê°€ ê°€ì‹œì£ ? 

ê·¸ ë‹¤ìŒì—ëŠ” **move_sido**ë¼ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ ì¤„ ê²ƒì´ì—ìš”. ì´ í•¨ìˆ˜ëŠ” ê´‘ì—­ì‹œë„ ì´ë¦„ì„ ë¦¬ìŠ¤íŠ¸ì— ì „ì†¡í•˜ê³  ê²€ìƒ‰ ë²„íŠ¼ì„ ëˆ„ë¥´ëŠ” ì—­í• ì„ í•´ìš”. 

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(driver, 10) # 10ì´ˆ ê¸°ë‹¤ë ¤ì£¼ì„¸ìš”! 

def move_sido(name):
    element = driver.find_element_by_id("cityCode") 
    element.send_keys(name)
    make_xpath = """//*[@id="searchBtn"]"""   # ê²€ìƒ‰ ë²„íŠ¼ì˜ xpath 
    wait.until(EC.element_to_be_clickable((By.XPATH,make_xpath)))  # ê²€ìƒ‰ ë²„íŠ¼ì´ ìƒê¸¸ ë•Œê¹Œì§€ ê¸°ë‹¤ë¦¬ê¸°! 
    driver.find_element_by_xpath(make_xpath).click() # ê²€ìƒ‰ ë²„íŠ¼ í´ë¦­ 
```

ì‹œë„ ë¦¬ìŠ¤íŠ¸ì—ì„œ ì›í•˜ëŠ” ì‹œë„ë¥¼ ê²€ìƒ‰í•  ë•Œ ëˆ„ë¥¼ ê³³ì˜ idë¥¼ ë¨¼ì € ì°¾ì•„ì£¼ê³ ìš”,("cityCode") ê·¸ ë‹¤ìŒì—ëŠ” ì•ì„œ ë½‘ì•„ë‚¸ ê´‘ì—­ì‹œë„ ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ë¦„ì— ë„£ì–´ì£¼ì–´ í´ë¦­í•˜ë„ë¡ í•  ê²ƒì´ì—ìš”. ì•„ë˜ ì‚¬ì§„ê³¼ ê°™ì´ìš”!

ê·¸ ë‹¤ìŒì—ëŠ” **append_data**ë¼ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ ê²ƒì¸ë°ìš”, ë¹ˆ ë‚´ìš©ìœ¼ë¡œ ë¯¸ë¦¬ ì¤€ë¹„í•œ ë”•ì…”ë„ˆë¦¬ì— append ëª…ë ¹ìœ¼ë¡œ ì½ì€ ë°ì´í„°ë¥¼ í•˜ë‚˜ì”© ì¶”ê°€í•˜ëŠ” ê¸°ëŠ¥ì´ì—ìš”. ì•ì˜ get_num í•¨ìˆ˜ë¥¼ í™œìš©í•´ì„œ ë¬¸ì¬ì¸, í™ì¤€í‘œ, ì•ˆì² ìˆ˜ í›„ë³´ë“¤ì˜ ë“í‘œìˆ˜ë§Œ ê°€ì ¸ì˜¬ ìˆ˜ ìˆë„ë¡ í•¨ìˆ˜ë¥¼ ë§Œë“  ê²ƒì´ì—ìš”.

```python
def append_data(df, sido_name, data):
    for each in df[0].values[1:]:
        data['ê´‘ì—­ì‹œë„'].append(sido_name)
        data['ì‹œêµ°'].append(each[0]) 
        data['pop'].append(each[2])
        data['moon'].append(get_num(each[3]))
        data['hong'].append(get_num(each[4]))
        data['ahn'].append(get_num(each[5]))
```

ë°ì´í„°ë¥¼ ë‹´ì•„ì„œ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë§Œë“¤ ê²ƒì´ë‹ˆ í•˜ë‚˜ ë¹ˆ ë³€ìˆ˜ë¥¼ ë§Œë“¤ì–´ ì£¼ê³ ìš”. 

```python
election_result_raw = {'ê´‘ì—­ì‹œë„' : [],
                       'ì‹œêµ°' : [],
                       'pop' : [],
                       'moon' : [],
                       'hong' : [],    
                       'ahn' : [] }
```

ì´ì œ ë°ì´í„°ë¥¼ ì½ì–´ë³¼ê¹Œìš©? ìš°ë¦¬ê°€ ì¶”ì¶œí•  ë°ì´í„°ëŠ” tableì´ë¼ëŠ” íƒœê·¸ì— ìˆì–´ìš”. ìš” íƒœê·¸ëŠ” í‘œë¥¼ ë‹´ëŠ” íƒœê·¸ì´ê±°ë“ ìš”. ì´ íƒœê·¸ë¥¼ ì½ì–´ì˜¤ë ¤ê³  í•©ë‹ˆë‹¤. 

- driver.page_source: ë¸Œë¼ìš°ì €ì— ë³´ì´ëŠ” ê·¸ëŒ€ë¡œì˜ HTML, í¬ë¡¬ ê°œë°œì ë„êµ¬ì˜ Element íƒ­ ë‚´ìš©ê³¼ ë™ì¼í•œ htmlì„ ê°€ì ¸ì˜¤ëŠ” ê²ƒ!
- [pandas.read_html](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_html.html): htmlì˜ table íƒœê·¸ë¥¼ pandasì˜ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ì „í™˜ì‹œì¼œì£¼ëŠ” ê²ƒ!

ìœ„ì˜ ê²ƒì„ ì‚¬ìš©í•´ì„œ ê° ì‹œë„ ì´ë¦„ì„ í´ë¦­í•´ì„œ tableì„ ê°€ì ¸ì™€ read_htmlì„ í†µí•´ ë°ì´í„°í”„ë ˆì„ í˜•íƒœë¡œ ì „í™˜ì‹œì¼œì¤€ë‹¤ê³  ë³´ë©´ ë˜ë„¤ìš”. 

```python
from bs4 import BeautifulSoup

for each_sido in sido_names_values:
    move_sido(each_sido)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table')  # tableì´ë€ íƒœê·¸ë¥¼ ì°¾ì•„ë¼! 
    
    df = pd.read_html(str(table))
    
    append_data(df, each_sido, election_result_raw)
```

ê´‘ì—­ì‹œê°€ ì•„ë‹Œë° í–‰ì •êµ¬ë¥¼ ê°€ì§€ê³  ìˆëŠ” ìˆ˜ì›, ì„±ë‚¨, ì•ˆì–‘ ë“±ì€ ì•ˆì–‘ ë§Œì•ˆ, ì•ˆì–‘ ë™ì•ˆê³¼ ê°™ì´ í‘œì‹œí•˜ê¸° ìœ„í•´ì„œ cut_char_siguë¼ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ ì¤ì‹œë‹¤. ë§Œì•½ì— ì´ë¦„ì˜ ê¸¸ì´ê°€ 2ë¥¼ ë„˜ì–´ê°€ë©´ ë’¤ì— 'êµ¬'ëŠ” ì‚­ì œí•˜ë„ë¡ í•˜ëŠ” ê±°ì£ . 

```python
# ì¤‘êµ¬, ë‚¨êµ¬ ê°™ì€ ë‘ê¸€ìëŠ” ê·¸ëŒ€ë¡œ, ì¤‘ë‘êµ¬ ì„œì´ˆêµ¬ ì˜ì™•ì‹œëŠ” ì¤‘ë‘, ì„œì´ˆ, ì˜ì™•ìœ¼ë¡œ ì¤„ì´ë ¤ê³  
def cut_char_sigu(name): 
    return name if len(name)==2 else name[:-1]
```

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75

    whitelabelmin = 20.

    datalabel = targetData

    tmp_max = max([ np.abs(min(blockedMap[targetData])), 
                                  np.abs(max(blockedMap[targetData]))])
    vmin, vmax = -tmp_max, tmp_max

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # ì§€ì—­ ì´ë¦„ í‘œì‹œ
    for idx, row in blockedMap.iterrows():
        # ê´‘ì—­ì‹œëŠ” êµ¬ ì´ë¦„ì´ ê²¹ì¹˜ëŠ” ê²½ìš°ê°€ ë§ì•„ì„œ ì‹œë‹¨ìœ„ ì´ë¦„ë„ ê°™ì´ í‘œì‹œí•œë‹¤. 
        #(ì¤‘êµ¬, ì„œêµ¬)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='ê³ ì„±':
            dispname = 'ê³ ì„±'
        else:
            dispname = row['ID']

        # ì„œëŒ€ë¬¸êµ¬, ì„œê·€í¬ì‹œ ê°™ì´ ì´ë¦„ì´ 3ì ì´ìƒì¸ ê²½ìš°ì— ì‘ì€ ê¸€ìë¡œ í‘œì‹œí•œë‹¤.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if np.abs(row[targetData]) > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # ì‹œë„ ê²½ê³„ ê·¸ë¦°ë‹¤.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
```
