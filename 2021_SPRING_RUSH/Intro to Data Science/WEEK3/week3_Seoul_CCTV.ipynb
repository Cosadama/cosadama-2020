책에는 이와 다르게 inplace 옵션을 사용하는데요, 저는 제 개인적인 취향은 inplace보다 이렇게 변수를 다시 정의해주는게 더 좋은 것 같더라구요. 본인이 원하는대로 하시면 됩니다.

#### 데이터 정렬하기 

CCTV가 어디에 많이 있고 없는지를 알고 싶어요. 그러면 '소계'가 CCTV 수니까 이를 정렬시켜볼게요.

그리고 강남 3구 중의 하나인 송파구가 CCTV가 가장 적은 구 중의 하나라는 것인데, 강남 3구면 송파구에는 예산이 다른 구보다는 있지 않을까?(정확히 몰라요, 걍 생각해보는 것이라 검증을 해봐야지요) 👉 CCTV를 많이 설치할 수 있을 텐데..? 등의 이런 생각을 해볼 수 있겠죠. 이렇게 생각해 보는 과정이 **탐색적 데이터 분석** 과정입니다. 

반대로 CCTV가 많이 설치되어 있는 TOP5 구들을 봅시다.

```python
CCTV_Seoul.sort_values(by='소계', ascending=False).head(5) 
```

상위 5구로는 강남구, 양천구, 서초구, 은평구, 용산구가 나오게 됩니다. 강남구와 서초구가 탑에 속해있으니, 내가 세운 가설이 맞을 수도 있겠군 등 이렇게 생각해보는 거죠..! 

#### 데이터 연산하기 

그 다음 살펴볼 것이 **3년간 CCTV 증가율**이에요. CCTV 데이터에 연도 데이터도 함께 있으니까 최근 증가율도 살펴볼 수 있어요. 
- (최근 3년간 CCTV 증가율) = (2014년 + 2015년 + 2016년) / (2013년도 이전)* 100 

```python
# 최근 3년간 CCTV 설치 증가율
CCTV_Seoul['최근증가율'] = ((CCTV_Seoul['2014년'] + CCTV_Seoul['2015년'] 
                       + CCTV_Seoul['2016년']) / CCTV_Seoul['2013년도 이전']) * 100
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

엇, 보니까 종로구가 정말 많이 증가했네요. 도봉구, 마포구도 그렇구요. 그럼 왜 이렇게 CCTV 설치율이 최근 사이에 늘어났을까도 고민해볼 수 있겠죠!

#### 결측치 처리하기

엇, 그런데 우리가 넘파이에서 배운 nan이 있어요. 이런 것을 결측치라고 합니다. **머신러닝을 돌릴 때나 데이터 분석을 할 때나 결측치가 있으면 예측이나 분석의 정확도가 떨어지고, 의도하지 않은 결과가 나올 수 있어요.** 그래서 이렇게 결측치가 있는 경우에는 NaN이 어디에 있는지를 알아서 **적절하게 처리**를 해줘야 해요.

```python
pop_Seoul[pop_Seoul['구별'].isnull()] 
```

```python
pop_Seoul.head()
```

아하, 이게 인덱스 26에만 있구나. 그리고 합계가 필요없어서 인덱스가 0, 26인 row를 지우려고 해요.
저는 결측치를 처리하는 방법으로 저는 이 데이터를 데이터에서 제거하기로 한거죠.

```python
pop_Seoul = pop_Seoul.drop([0, 26])
pop_Seoul.head(25)
```

다시 변수를 정의해주었습니다. 이렇게 되니까 잘 제거되었는지 보고싶죠? 

```python
pop_Seoul[pop_Seoul['구별'].isnull()] 
```

오, 없네요! 결측치를 잘 제거했습니다.

#### 데이터 탐색하기 

이 작가는 고령자비율과 외국인비율을 계산했어요. 고령자비율과 외국인비율이 CCTV 설치율과 영향이 있는가 보고 싶은가봐요. 그러면 우리도 한번 고령자비율과 외국인 비율을 계산해볼게요!

```python
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100 #외국인비율 칼럼 생성
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100 #고령자비율 칼럼 생성
pop_Seoul.head()
```

이렇게 하면 데이터를 잘 알수 없으니까 또 오름차순과 내림차순을 해서 알아보도록 하죠.

```python
pop_Seoul.sort_values(by='인구수', ascending=False).head(5)
```

인구수를 기준으로 내림차순을 해주면, 송파구가 인구가 제일 많네요!

#### 상관계수 분석

우리는 간단한 상관계수 분석을 해볼거에요. 상관관계는 관계를 알고자 하는 두 변수가 모두 연속변수일 경우에만 계산이 가능하고, 이는 인과관계를 분석하는게 아닌 두 연속변수가 관련이 있는지 없는지를 알아보는 기술통계 분석이에요. 두 변수를 x,y라고 하면 상관관계 계수(coefficient)는 x값과 y값을 모두 z값으로 표준화해서 Zx * Zy를 모두 더한다음, n(총 개수)으로 나눠준 것이 상관계수에요. 다른 것은 다됐고 여기서 상관계수는 넘파이가 다 계산해 줄 테니까 우리는 여기서 **상관계수가 무엇을 의미**하는지만 알면 돼요!

- 0 ~ 0.2 : 매우 약한 상관관계 
- 0.2 ~ 0.4 : 약한 상관관계 
- 0.4 ~ 0.6 : 어느 정도의 상관관계 
- 0.6 ~ 0.8 : 강한 상관관계 
- 0.8 ~ 1.0 : 매우 강한 상관관계 

상관계수의 범위는 -1부터 1까지 음수가 나오면 음의 상관관계, 양수가 나오면 양의 상관관계라는 것까지만 알면 됩니다. 

그럼 다수의 데이터 중 상관계수가 가장 큰 값은 데이터를 비교해볼게요. **numpy의 corrcoef 명령**을 사용하면 상관계수를 알아서 계산해줍니다. 

```python
# 약한 음의 상관관계 
np.corrcoef(data_result['고령자비율'], data_result['소계'])
```

```python
# 상관관계가 거의 없다고 할 수 있음. 
np.corrcoef(data_result['외국인비율'], data_result['소계'])
```

```python
# 0.3이어서 약한 상관관계 
np.corrcoef(data_result['인구수'], data_result['소계'])
```

##### 추세선 그리기 - 선형 회귀선
(사회통계론을 듣지 않았거나 이 부분을 읽어도 이해가 가지 않는다면, 코드만 복붙해서 실행해보고 넘어가면 됩니다!)

오, 이제는 여기에 대충 경향을 보고 싶어요. 아까 상관계수가 어느 정도 약하게 있다고 나왔잖아요. 그래서 요 데이터를 가지고 경향성을 나타낼 수 있는 하나의 직선을 만들어줄 겁니다. 우리가 넘파이 실습과제에서 하나의 선을 그렸듯이 넘파이의 **polyfit 함수**를 사용해서 만들어줄 거에요. polyfit()은 **n차 다항식으로 회귀분석을 할 때 사용하는 함수**에요.

- np.polyfit(x, y, n): x는 x에 들어갈 데이터, y는 y에 들어갈 데이터, n는 다항식의 차수

우리가 그리고 싶은 추세선은 직선이라 y=ax+b라고 하죠. 직선은 1차원이니까 n=1이고, polyfit의 결과값은 [a,b]가 나오게 되어요. 즉, 기울기와 절편값이 나오는거죠. 우리가 사통론에서 요 a,b값을 구하려고, 일일이 하나씩 다 계산해서 정말 힘들게 계산했잖아요. 그런데 이렇게 polyfit()을 사용하면 바로 기울기와 절편값을 줍니다. 2,3,4 등 n차원까지 모두 가능해요.

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'],1)
fp1
```

그럼 이건 주어진 데이터를 가지고 만들어낸 1차 방정식의 기울기와 절편값이에요. 그 다음에는 **poly1d()**를 가지고 직선을 만들어볼게요. 

- [poly1d()란?](https://docs.scipy.org/doc/numpy/reference/generated/numpy.poly1d.html)

이것을 보시고 이해를 해보세요. poly1d는 다항식(polynomial function)을 정의해주는 함수에요. 즉, 우리가 poly1d() 안에 [a,b]를 넣는다면, ax+b라는 1차 방정식을 넣어주는거죠. 만약 poly1d([1,2,3])이라면 x^2+2x+3을 만들어주는 것이구요. 

```python
f1 = np.poly1d(fp1)
```

f1 자체가 방정식이니까, 직선을 만드려면 x값을 공급해줘야해요. 그 x값은 **np.linspace**로 만들어 줄 거에요. 

- np.linspace(start, stop, num): num은 데이터 개수

np.linspace 함수는 start부터 stop의 범위에서 num개를 균일한 간격으로 데이터를 생성하고 배열을 만드는 함수에요. 

```python
fx = np.linspace(100000, 700000, 100)
```

그러니까 정리를 하면, x에는 fx를 공급해주고, y로는 다항식을 만들어 준 f1을 넣어주는 것이죠. 그래서 점선의 일차원 그래프를 만들어보면! 

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

이렇게 추세선 완성! 오, CCTV와 인구수에 이런 추세선이 그려지는 군요. 

여기에다가 저자는 두 가지 장치를 더했어요. 하나는 직선의 경향에서 멀어진 구일수록 색이 변한다는 것과, 다른 하나는 경향에서 멀어진 구의 이름을 표시하고 싶다는 거에요. 그래서 이를 표시하기위해 오차를 계산하는 코드를 만들었어요. 

- 오차 = |(CCTV수) - f1(인구수)|

엇, 이거 우리 회귀분석할 때 선형 회귀식을 만들 때 봤죠? 아까 우리가 추세선을 만들었잖아요. 그 추세선이 바로 **회귀식**이랍니다. **오차는 회귀선을 통해서 나온 y값과 실제 y값의 차이**잖아요. 그래서 요거를 이용해서 오차를 계산해 data_result에다 새로운 컬럼으로 넣어주어봤습니다.

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'],1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

# 오차 칼럼 만들어주기 
data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))
data_result.head()
```

그럼 이것을 '오차'를 기준으로 데이터를 정렬해보도록 할게요. 

```python
df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

그럼 이제 텍스트와 color map을 입혀볼게요. 

- color는 plt.scatter()의 c 옵션값으로 넣어줬어요. 
- plt.text(x, y, s): x와 y는 (x,y)위치(the position to place the text), s는 넣어줄 텍스트(the text)

우리 파이썬을 배웠기 때문에 아래의 for문 이해할 수 있어요💖 

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['소계'],
           c=data_result['오차'], s=50)                   #c=marker color, s=marker size
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')       #lw=line width

# 텍스트 만들어주기
for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,   
            df_sort.index[n], fontsize=15)

plt.xlabel('인구수')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```

df_sort는 오차가 높은 순으로 배열을 해놓았죠. 그래서 for문을 사용해 range(10), 즉 상위 10개만 text를 보이도록 해줬어요. 

- df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98

이렇게 적은 이유는 점의 바로 위에 이름이 올라가게 하지 말고, 오른쪽으로 이름이 위치할 수 있도록 한거에요. 이름이 위치한 곳이 모두 다 오른쪽 살짝 옆에 있죠? 이것을 for문으로 구현한 것이랍니다.

### 2.5 분석 결과 

직선을 기준으로 위에 있는 '강남구', '양천구', '서초구', '은평구', '용산구'는 서울시 전체 지역의 일반적인 경향보다 CCTV가 많이 설치된 지역이네요. 그리고 '송파구','강서구','중랑구','마포구','도봉구'는 일반적인 경향보다 CCTV가 적게 설치된 지역이고요. 특히 '강남구'는 월등히 많은 CCTV가 설치됐지만, '송파구'는 인구수에 비해 너무나도 적은 수의 CCTV를 가지고 있네요!

여기다 이제 우리는 사회학하는 친구들이니까 왜 이렇게 될까를 생각해봐야겠죠. 앞서 처음에 생각했던 '구별 예산 정도에 따라 CCTV 설치수가 다를꺼다'라는 것을 증명할 수는 없지만(예산 데이터가 없으니까요), 왜 저렇게 강남구는 아웃라이어로 있나라고 생각해볼 수 있겠죠 😊
