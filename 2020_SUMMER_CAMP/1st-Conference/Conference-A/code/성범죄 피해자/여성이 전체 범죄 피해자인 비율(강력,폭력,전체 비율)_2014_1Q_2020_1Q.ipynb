```python
import pandas as pd
import numpy as np
```

```python
female_victim_raw = pd.read_csv('여성피해자+발생건수+및+비율.csv', thousands=',', encoding='utf-8')
female_victim_raw
```

```python
female_victim_raw.info()
```

```python
female_victim_raw = female_victim_raw.drop(['No'], 1)
female_victim_raw
```

```python
female_victim_raw = female_victim_raw.drop([0])
female_victim_raw
```

```python
female_victim_raw['건수'] = female_victim_raw['건수'].str.replace(',','')
female_victim_raw['건수'] = female_victim_raw['건수'].astype(int)
female_victim_raw
```

```python
female_victim_raw['비율'] = female_victim_raw['비율'].str.replace(',','')
female_victim_raw['비율'] = female_victim_raw['비율'].astype(float)
female_victim_raw
```

```python
female_victim_all = female_victim_raw[0::3]
female_victim_all
```

```python
# 강력범죄니까 strong..?
female_victim_strong = female_victim_raw[1::3]
female_victim_strong
```

```python
female_victim_violence = female_victim_raw[2::3]
female_victim_violence
```

```python
from datetime import datetime
from datetime import timedelta #timedelta를 더하거나 뺄 수 있음.
from dateutil.parser import parse
```

```python
rng = pd.period_range('2014Q1', '2020Q1', freq='Q-JAN')
rng
```

```python
female_victim_all['기간'] = rng
female_victim_strong['기간'] = rng
female_victim_violence['기간'] = rng
```

```python
female_victim_all.head()
```

```python
female_victim_all = female_victim_all.drop(['자료시점'], 1)
female_victim_all = female_victim_all.drop(['분류'], 1)

female_victim_strong = female_victim_strong.drop(['자료시점'], 1)
female_victim_strong = female_victim_strong.drop(['분류'], 1)

female_victim_violence = female_victim_violence.drop(['자료시점'], 1)
female_victim_violence = female_victim_violence.drop(['분류'], 1)
```

```python
female_victim_all.info()
female_victim_strong.info()
female_victim_violence.info()
```

```python
female_victim_all.set_index('기간', inplace=True)
female_victim_strong.set_index('기간', inplace=True)
female_victim_violence.set_index('기간', inplace=True)
```
