μ κ·Έλ¬λ©΄ λ¬Έμ  μƒν™©μ„ μ–΄λ–»κ² ν•΄κ²°ν•λ©΄ μΆ‹μ„κΉμ”? μ°λ¦¬λ” [fillna](https://rfriend.tistory.com/262) λ©”μ„λ“λ¥Ό ν™μ©ν• κ±°μ—μ”. fillnaλ¥Ό ν™μ©ν•μ—¬ λΉ μ…€μ—λ” κ·Έ μ•μ λ‚΄μ©μ„ μ±„μ°λ„λ΅ ν• κ²λ‹λ‹¤.

λν• ν–‰μ •κµ¬μ—­ μΉΌλΌ μ΄λ¦„κ³Ό κ³„λ„ λ” μ§κ΄€μ μΌλ΅ μ΄ν•΄ν•κΈ° μ‰½κ² λ°”κΏ”μ¤„κ±°μ—μ”.

```python
population.fillna(method='pad', inplace=True)
population.rename(columns = {'ν–‰μ •κµ¬μ—­(λ™μλ©΄)λ³„(1)':'κ΄‘μ—­μ‹λ„', 
                             'ν–‰μ •κµ¬μ—­(λ™μλ©΄)λ³„(2)':'μ‹λ„', 
                             'κ³„':'μΈκµ¬μ'}, inplace=True)

population = population[(population['μ‹λ„'] != 'μ†κ³„')]

population.head()
```

ν•­λ©μ„ κµ¬λ¶„μΌλ΅ λ°”κΎΈκ³ , μ΄μΈκµ¬μ, μ—¬μμΈκµ¬μ μ΄λ¦„λ„ μ§κ΄€μ μΌλ΅ μ΄ν•΄ν•κΈ° μ‰½κ²λ” λ°”κΏ”μ¤„κ±°μ—μ”.

```python
population.is_copy = False  # μ™ λ„£λ”μ§€λ” μ•„λμ λ‹µλ³€ μ°Έκ³ 

population.rename(columns={'ν•­λ©':'κµ¬λ¶„'}, inplace=True)   # ν•­λ© -> κµ¬λ¶„

population.loc[population['κµ¬λ¶„']=='μ΄μΈκµ¬μ (λ…)', 'κµ¬λ¶„'] = 'ν•©κ³„'    # μ΄μΈκµ¬μ -> ν•©κ³„
population.loc[population['κµ¬λ¶„']=='λ‚¨μμΈκµ¬μ (λ…)', 'κµ¬λ¶„'] = 'λ‚¨μ'  # λ‚¨μμΈκµ¬μ -> λ‚¨μ
population.loc[population['κµ¬λ¶„']=='μ—¬μμΈκµ¬μ (λ…)', 'κµ¬λ¶„'] = 'μ—¬μ'  # μ—¬μμΈκµ¬μ -> μ—¬μ

population.head()
```

### 7.3 μΈκµ¬ μ†λ©Έ μ„κΈ° μ§€μ—­ κ³„μ‚°ν•κ³  λ°μ΄ν„° μ •λ¦¬ν•κΈ°  
μΈκµ¬ μ†λ©Έ μ„κΈ° μ§€μ—­μ„ μ•κΈ° μ„ν•΄μ„  **65μ„Έ μ΄μƒ λ…ΈμΈ μΈκµ¬**μ™€ **20-39μ„Έ μ—¬μ„± μΈκµ¬**λ¥Ό μ•μ•„μ•Ό ν•λ‹¤κ³  ν–μµλ‹λ‹¤. μ§€κΈμ λ°μ΄ν„° κµ¬λ¶„ λ°©μ‹μ€ λ§μ§€ μ•μΌλ‹ ν•©ν•  κ±΄ ν•©ν•κ³  μ¶”κ°€ν•  κ±΄ μ¶”κ°€ν•΄λ΄…μ‹λ‹¤!

```python
population['20-39μ„Έ'] = population['20 - 24μ„Έ'] + population['25 - 29μ„Έ'] + \
                        population['30 - 34μ„Έ'] + population['35 - 39μ„Έ']
    
population['65μ„Έμ΄μƒ'] = population['65 - 69μ„Έ'] + population['70 - 74μ„Έ'] + \
                        population['75 - 79μ„Έ'] + population['80 - 84μ„Έ'] + \
                        population['85 - 89μ„Έ'] + population['90 - 94μ„Έ'] + \
                        population['95 - 99μ„Έ'] + population['100+']
            
population.head(10)
```

μ΄μ  ν•„μ”ν• λ°μ΄ν„°λ“¤μ€ μ–Όμ¶” λ§λ“¤μ–΄μ΅μµλ‹λ‹¤. λ°μ΄ν„°λ“¤μ„ μ •λ¦¬ν•λ©΄ μ΅°κΈ λ” λ³΄κΈ° νΈν•  κ²ƒ κ°™μ•„μ”. piviot_tableμ„ μ΄μ©ν•΄ λ°μ΄ν„°λ“¤μ„ μ μ •λ¦¬ν•΄μ¤μ‹λ‹¤.

```python
pop = pd.pivot_table(population, 
                     index = ['κ΄‘μ—­μ‹λ„', 'μ‹λ„'], 
                     columns = ['κµ¬λ¶„'],
                     values = ['μΈκµ¬μ', '20-39μ„Έ', '65μ„Έμ΄μƒ'])
pop.head()
```

μ΄λ ‡κ² μ»¬λΌ 2μΈµ, μΈλ±μ¤ 2κ°μΈ λ°μ΄ν„°ν”„λ μ„μ΄ λ§λ“¤μ–΄μ΅μµλ‹λ‹¤. μ΄μ λ” μ—¬κΈ°μ—λ‹¤κ°€ μ†λ©ΈλΉ„μ¨μ„ κ³„μ‚°ν•λ©΄ λκ² μ£ ! 20-39μ„Έ μ—¬μ„±μΈκµ¬μ™€ 65μ„Έμ΄μƒ ν•©κ³„λ¥Ό μ‚¬μ©ν•΄μ„ κ³„μ‚°ν•  μ μμµλ‹λ‹¤.

```python
pop['μ†λ©ΈλΉ„μ¨'] = pop['20-39μ„Έ','μ—¬μ'] / (pop['65μ„Έμ΄μƒ','ν•©κ³„'] / 2)
pop.head()
```

μ½”λ“ ν• μ¤„λ΅ μ΄λ ‡κ² λ§μ€ μ—°μ‚°μ΄ λλ‹¤λ‹... μ‹ κΈ°ν•μ§€ μ•λ‚μ”?   
κ·ΈλΌ μ†λ©Έ μ„κΈ° μ§€μ—­μ΄ μ–΄λ””μΈμ§€λ¥Ό μƒλ΅μ΄ μΉΌλΌμ„ λ§λ“¤μ–΄ λ„£μ–΄μ£Όκ² μµλ‹λ‹¤

```python
pop['μ†λ©Έμ„κΈ°μ§€μ—­'] = pop['μ†λ©ΈλΉ„μ¨'] < 1.0  # BooleanμΌλ΅ ν‘μ‹ν•΄μ¤„κ±°μ—μ”(T or F)
pop.head()
```

κ·Έλ°λ° μ΄λ ‡κ² μ •λ¦¬ν•  κ²½μ°μ— μΈλ±μ‹±ν•κΈ°κ°€ κµ‰μ¥ν κΉλ‹¤λ΅μ›μ§„λ‹¤λ” λ‹¨μ μ΄ μμµλ‹λ‹¤. μλ¥Ό λ“¤μ–΄ **κ°•μ›λ„ κ°•λ¦‰μ‹μ 20~39μ„Έ μ—¬μ„± μΈκµ¬**λ¥Ό μ•κ³  μ‹¶μΌλ©΄ μ…λ ¥ν•΄μ•Όν•λ” μ΅°κ±΄ μΉΌλΌ μ΅°κ±΄ 2κ°(20~39μ„Έ, μ—¬μ), λ΅μ°μ—μ„ μ΅°κ±΄ 2κ°(κ°•μ›λ„, κ°•λ¦‰μ‹)κ°€ ν•„μ”ν•©λ‹λ‹¤. μ΄λ ‡κ² λλ©΄ ν›„μ— μ§€λ„ μ‹κ°ν™”λ¥Ό ν•  λ• μ‘μ„±ν•΄μ•Ό ν•λ” μ½”λ“κ°€ λ„λ¬΄ λ³µμ΅ν•΄μ Έλ²„λ¦¬κ² μ£ ?

μ°μ„  reset_index()λ¥Ό ν•΄μ£Όκ³ μ”

```python
pop.reset_index(inplace=True) 
pop.head()
```

```python
# νμ‹λ‹ μ½”λ“: drawKorea()λΌλ” ν•¨μλ¥Ό λ§λ“¤μ–΄ μ¤€κ²ƒ! 
def drawKorea(targetData, blockedMap, cmapname):  # μ—¬κΈ° μ£Όμν•΄μ„ λ³΄κΈ°!
    gamma = 0.75
    
    # μ‹κ°ν™” μ‹μ— μ†λ©Έμ§€μ—­μ΄ μ•„λ‹ κ³³μ€ ν°μƒ‰μΌλ΅ ν‘μ‹λκ²λ”! 
    whitelabelmin = (max(blockedMap[targetData]) - 
                                     min(blockedMap[targetData]))*0.25 + \
                                                                min(blockedMap[targetData])

    datalabel = targetData

    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # μ§€μ—­ μ΄λ¦„ ν‘μ‹
    for idx, row in blockedMap.iterrows():
        # κ΄‘μ—­μ‹λ” κµ¬ μ΄λ¦„μ΄ κ²ΉμΉλ” κ²½μ°κ°€ λ§μ•„μ„ μ‹λ‹¨μ„ μ΄λ¦„λ„ κ°™μ΄ ν‘μ‹ν•λ‹¤. 
        #(μ¤‘κµ¬, μ„κµ¬)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='κ³ μ„±':
            dispname = 'κ³ μ„±'
        else:
            dispname = row['ID']

        # μ„λ€λ¬Έκµ¬, μ„κ·€ν¬μ‹ κ°™μ΄ μ΄λ¦„μ΄ 3μ μ΄μƒμΈ κ²½μ°μ— μ‘μ€ κΈ€μλ΅ ν‘μ‹ν•λ‹¤.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # μ‹λ„ κ²½κ³„ κ·Έλ¦°λ‹¤.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
```

### 7.8 μΈκµ¬ ν„ν™©μ—μ„ μ—¬μ„± μΈκµ¬ λΉ„μ¨ ν™•μΈν•κΈ°

(κ·Όλ° μ–΄μ°¨ν”Ό λ‹¤ κ²€μ€μƒ‰μΌλ΅ λ‚μ¤κΈ° λ•λ¬Έμ— μƒ‰μ€ μ¤‘μ”ν•μ§€ μ•μ§€λ§, λ°μ΄ν„°λ¥Ό μ† λ³΄λ©΄ μμκ°€ λ° μ μμ–΄μ„ μ¤λ¥κ°€ λ‚μ„ μμ •ν• κ±°μ—μ”)

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75

    whitelabelmin = 20.

    datalabel = targetData
    
    # μμκ°€ μƒκΈ°μ§€ μ•λ„λ΅! 
    tmp_max = max([ np.abs(min(blockedMap[targetData])), 
                                  np.abs(max(blockedMap[targetData]))])
    vmin, vmax = -tmp_max, tmp_max

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # μ§€μ—­ μ΄λ¦„ ν‘μ‹
    for idx, row in blockedMap.iterrows():
        # κ΄‘μ—­μ‹λ” κµ¬ μ΄λ¦„μ΄ κ²ΉμΉλ” κ²½μ°κ°€ λ§μ•„μ„ μ‹λ‹¨μ„ μ΄λ¦„λ„ κ°™μ΄ ν‘μ‹ν•λ‹¤. 
        #(μ¤‘κµ¬, μ„κµ¬)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='κ³ μ„±':
            dispname = 'κ³ μ„±'
        else:
            dispname = row['ID']

        # μ„λ€λ¬Έκµ¬, μ„κ·€ν¬μ‹ κ°™μ΄ μ΄λ¦„μ΄ 3μ μ΄μƒμΈ κ²½μ°μ— μ‘μ€ κΈ€μλ΅ ν‘μ‹ν•λ‹¤.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if np.abs(row[targetData]) > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # μ‹λ„ κ²½κ³„ κ·Έλ¦°λ‹¤.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
```
