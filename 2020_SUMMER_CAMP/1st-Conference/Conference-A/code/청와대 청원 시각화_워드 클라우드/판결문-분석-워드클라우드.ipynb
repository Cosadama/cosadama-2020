```python
def reader(pdf):
    path = "D:\\판결문\\" + pdf
    with open(path, "rb") as f:
        pdf = pdftotext.PDF(f)
    total=[]
    for page in pdf:
        total.append(page)
    raw = str(total)
    if (raw.find("양형의 이유") != -1) and (raw.find("신상정보 등록") != -1):
        raw = raw[raw.find("양형의 이유"):raw.find("신상정보 등록")]
        print('완료')
    elif (raw.find("양형의 이유") != -1) and (raw.find("신상정보 등록") == -1):
        raw = raw[raw.find("양형의 이유") :]
        print('완료')
    else:
        print("양형의 이유 없어요!")
        
    return raw
```
