_"부유할수록 CCTV개수가 많은 경향을 보이는구나. 그런데 왜 강남 3구라고 불리는 송파구는 CCTV가 적을까? 범죄가 적어서? 인구가 적어서는 아닌거 같은데? 예산이 다른데 배정되어있나?"_

이렇습니다.

이렇게 생각해보는 과정을 **탐색적 데이터 분석**이라고 합니다. 여기서부터 문제의식이 구체화되고, 분석할 범위가 좁혀지니 되게되게 중요하겠죠?

#### 데이터 연산하기

이번에는 **3년간 CCTV가 어떻게 증가했는지** 살펴보면서 데이터를 연산해봅시다. 이렇게 시간순서대로 정렬된 데이터는 증가율이나 감소율처럼 추세를 볼 수 있습니다.
 - (최근 3년간 증가율) = (2014년 + 2015년 + 2016년) / (2013년 이전) * 100

```python
# 최근 3년간 CCTV 설치 증가율
CCTV_Seoul['최근증가율'] = ((CCTV_Seoul['2014년'] + CCTV_Seoul['2015년'] 
                       + CCTV_Seoul['2016년']) / CCTV_Seoul['2013년도 이전']) * 100
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

오! 종로구, 도봉구, 마포구는 아까 하위 5개구에 있던 곳이죠? 3년 사이에 최고로 많이 늘었는데도 하위권이네요? 이번에는 

**이 구들은 CCTV개수가 다른 구에 비해 엄청 부족했거나, 다른 구들의 CCTV 설치 속도가 완만해졌지 않을까?**

#### 데이터 탐색하기

#### 상관계수 분석

간단하게 아까 눈으로 했던 상관관계를 분석해봅시다. 상관관계는 알고자 하는 두 변수가 모두 **연속변수**일 때 계산이 가능합니다.(중간에 끊기면 안됨) 상관관계를 분석하면 이 **두 연속변수가 서로 관련이 있는지 없는지 알 수 있습니다.** **인과관계는 아니라는거~**(인과는 원인과 결과)

두 변수를 x,y라고 하면 상관관계 계수(coefficient)는 x값과 y값을 모두 z값으로 표준화해서 Zx * Zy를 모두 더한다음, n(총 개수)으로 나눠준 것이 상관계수에요.

오 무슨말인지 잘 모르겠다구요? 그럼 김보겸씨에게 물어보도록 합시다!! 지금 당장 깊이있게 알지 않아도 괜찮습니다. 다른 것은 다됐고 여기서 상관계수는 넘파이가 다 계산해 줄 테니까 우리는 여기서 **상관계수가 무엇을 의미하는지**만 알면 됩니다!

 - 0 ~ 0.2 : 매우 약한 상관관계
 - 0.2 ~ 0.4 : 약한 상관관계
 - 0.4 ~ 0.6 : 어느 정도의 상관관계
 - 0.6 ~ 0.8 : 강한 상관관계
 - 0.8 ~ 1.0 : 매우 강한 상관관계
 
상관관계는 -1부터 1까지 수로 표현됩니다. 음수는 음의 상관관계, 양수는 양의 상관관계죠. 음수면 내가 생각한 가설에 반대결과고, 양수면 가설에 부합한다고 생각하시면 됩니다.
 
그럼 다수의 데이터 중 상관계수가 가장 큰 값은 데이터를 비교해볼게요. **numpy의 corrcoef** 명령을 사용하면 상관계수를 알아서 계산해줍니다.

```python
# 약한 음의 상관관계 
np.corrcoef(data_result['고령자비율'], data_result['소계'])
```

```python
# 상관관계가 거의 없다고 할 수 있음. 
np.corrcoef(data_result['외국인비율'], data_result['소계'])
```

```python
# 0.3이어서 약한 상관관계 
np.corrcoef(data_result['인구수'], data_result['소계'])
```

외국인 비율은 거의 상관관계가 없고, '인구수'가 그나마 유의미하네요. 그럼 CCTV와 인구수의 관계를 더 파봅시다.

#### 추세선 그리기 - 선형 회귀선

(사회통계론을 듣지 않았거나 이 부분을 읽어도 이해가 가지 않는다면, 코드만 복붙해서 실행해보고 넘어가면 됩니다!)

오, 이제는 여기에 대충 경향을 보고 싶어요. 아까 상관계수가 어느 정도 약하게 있다고 나왔잖아요. 그래서 요 데이터를 가지고 경향성을 나타낼 수 있는 하나의 직선을 만들어줄 겁니다. 넘파이의 polyfit 함수를 사용해서 만들어줄 거에요. **polyfit()은 n차 다항식으로 회귀분석을 할 때 사용하는 함수**입니다.

**np.polyfit(x, y, n): x는 x에 들어갈 데이터, y는 y에 들어갈 데이터, n는 다항식의 차수.**
우리가 그리고 싶은 추세선은 y=x^2같은 곡선이 아니고 직선이라 y=ax+b라고 하죠.

직선은 1차원이니까 n=1이고, polyfit의 결과값은 [a,b]가 나오게 됩니다. 즉, 기울기와 절편값이 나오는거죠. 

우리가 사통론이나 통계학 개론에서 요 a,b값을 구하려고, 일일이 하나씩 다 계산해서 정말 힘들게 계산했잖아요.(그러다가 저는 영 좋지 못한 성적을 받았습니다^^) 그런데 이렇게 polyfit()을 사용하면 바로 기울기와 절편값을 줍니다. 2,3,4 등 n차원까지 모두 가능해요. 진짜 저처럼 계산 싫어하는 사람들한테 최고... 이런 통계적 기술들이 어떤 의미를 갖는지(+ 어떤 과정에서 도출되었는지)만 알면 컴퓨터가 알아서 다 해줍니다ㅠㅠㅠㅠ

 그래도 **회귀분석**에 대해 간단하게 설명해 드리겠습니다. 위의 산점도에 찍힌 점들은 각각 y값을 갖습니다. 이 y값의 차이(잔차residual, 오차error)가 최소가 되는 직선을 그리는거예요. 이해가 안된다면 산점도에다가 일단 임의의 직선을 하나 긋고, x값에 해당하는 원래 데이터의 값(산점도의 점)과 직선y=ax+b 사이의 차이(의 제곱-음수가 나올 수 있기 때문)가 최소가 되게하는 기울기a와 절편b 값으로 옮겨준다고 생각하시면 됩니다. 

 이렇게 회귀분석을 하면 두 변수 사이의 **인과관계**를 통계적으로 추리할 수 있습니다. 종속변수가 독립변수에 어떤 영향을 받는지 알 수 있죠
 
 (재미있는 사실은 프랜시스 갈튼Francis Galton이라는 우생학자-어떤 종이 더 우월하다고 주장하여 인종차별의 근거가 된 이론-가 만든 분석기법이라고 합니다.^^)

p.s. 나중에 머신러닝을 공부하면 여러가지 회귀분석을 배우게 됩니다. Linear Regression, Ridge, Lasso 등등. 머신러닝에서는 앞에서 이야기 한 기울기a와 절편값b을 각각 coefficient(coef), intercept라고 하더라구요. 코드도 말할 때도 그냥 영어로 저렇게 쓰는 것 같아요.

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'],1)
fp1
```

이 결과는 주어진 데이터를 가지고 만들어낸 1차 방정식의 기울기a(coef)와 절편값b(intercept)입니다. 1.3091어쩌구가 기울기, 6.45066뭐시기가 절편값이겠죠?

그 다음에는 [poly1d()](https://docs.scipy.org/doc/numpy/reference/generated/numpy.poly1d.html) 를 가지고 직선을 만들어볼게요.

위에 링크를 보시고 이해를 해보세요. poly1d는 다항식(polynomial function)을 정의해주는 함수입니다. 즉, 우리가 poly1d() 안에 [a,b]를 넣는다면, ax+b라는 1차 방정식을 넣어주는거죠. 만약 poly1d([1,2,3])이라면 x^2+2x+3을 만들어주는 것이구요.

```python
f1 = np.poly1d(fp1)
f1
```

- x값(정의역): f1 자체가 방정식이니까, 직선을 만드려면 x값(정의역)을 공급해줘야해요. 그 x값은 **np.linspace**로 만들어 줄 거에요.

**np.linspace(start, stop, num): num은 데이터 개수**
np.linspace 함수는 start부터 stop의 범위에서 num개를 균일한 간격으로 데이터를 생성하고 배열을 만드는 함수에요.

- y값(공역): f1에 정의역을 넣은 값들. 우리가 따로 계산해 주지 않아도 됨. 왜냐하면 f1이라는 방정식(함수)가 이미 있기 때문!!

```python
fx = np.linspace(100000, 700000, 100)
```

정리하자면 
1. fp1 = np.polyfit()으로 구하려는 함수의 기울기a(coef)와 절편b(intercept)을 알아낸다 -> 1차 함수(방정식) 도출

2. np.poly1d()의 괄호 안에 fp1을 넣어서 a를 기울기로 하고, b를 절편으로 하는 다항식을 만들어 준다. 

3. 이때, 이 함수의 정의역이 정해져야 하므로 np.linespace()로 어디서부터 어디까지 그릴지 설정해 준다.

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

추세선 완성! 오, CCTV와 인구수에 이런 추세선이 그려지네요. 갑자기 사회조사방법론에서 어떤 친구가 임의로 추세선을 그렸다가 걸린게 생각납니다^^.

여기에다가 저자는 두 가지 장치를 더했어요. 
1. **직선의 경향에서 멀어진 구일수록 색을 변화시키기**

2. **경향에서 멀어진 구의 이름을 표시하기**

그래서 이를 표시하기위해 오차를 계산하는 코드를 만들었어요.

오차 = |(CCTV수) - f1(인구수)|

아까 우리가 추세선을 만들었는데, 사실 그 **추세선이 바로 회귀식**이랍니다!! 

**오차는 회귀선을 통해서 나온 y값과 실제 y값의 차이**잖아요. 그래서 요거를 이용해서 오차를 계산해 data_result에다 새로운 컬럼으로 넣어주어봤습니다.

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'],1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

# 오차 칼럼 만들어주기 
data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))
data_result.head()
```

```python
# 오차 기준으로 정렬
df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

그럼 이제 텍스트와 color map을 표현해 봅시다.

 - color는 plt.scatter()의 c 옵션값으로 넣어주면 되고,
 - plt.text(x, y, s): x와 y는 (x,y)위치(the position to place the text), s는 넣어줄 텍스트(the text)
 
for문으로 한번에 처리 해버리기~

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['소계'],
           c=data_result['오차'], s=50)                   #c=marker color, s=marker size
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')       #lw=line width

# 텍스트 만들어주기
for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98, 
            df_sort.index[n], fontsize=15)

plt.xlabel('인구수')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```

for문 안에 1.02나 0.98은 찍힌 점으로부터 어떤 위치에 표시할지 정해주는 수치예요. 점에 글자가 가리면 안보이니까 이렇게 처리해 주는거랍니다. 그래프를 보면 점의 오른쪽 살짝 위에 위치한 것을 볼 수 있죠?
