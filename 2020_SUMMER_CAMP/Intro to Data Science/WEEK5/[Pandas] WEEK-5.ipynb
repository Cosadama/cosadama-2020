다른 곳은 다 괜찮은데 행정구역 칼럼에만 NaN 값이 있네요! 이럴 때는 원데이터셋을 한 번 보는게 좋아요. 엑셀 파일을 열어보면 행정구역 칼럼이 병합되어 있어요. 이 병합이 데이터프레임으로 왔을 때 문제를 일으키고 있는 것입니다. 

자 그러면 문제 상황을 어떻게 해결하면 좋을까요? 우리는 [fillna](https://rfriend.tistory.com/262) 메서드를 활용할거에요. fillna를 활용하여 빈 셀에는 그 앞의 내용을 채우도록 할겁니다.

또한 행정구역 칼럼 이름과 계도 더 직관적으로 이해하기 쉽게 바꿔줄거에요.

```python
population.fillna(method='pad', inplace=True)
population.rename(columns = {'행정구역(동읍면)별(1)':'광역시도', 
                             '행정구역(동읍면)별(2)':'시도', 
                             '계':'인구수'}, inplace=True)

population = population[(population['시도'] != '소계')]

population.head()
```

항목을 구분으로 바꾸고, 총인구수, 여자인구수 이름도 직관적으로 이해하기 쉽게끔 바꿔줄거에요.

```python
population.is_copy = False  # 왜 넣는지는 아래의 답변 참고

population.rename(columns={'항목':'구분'}, inplace=True)   # 항목 -> 구분

population.loc[population['구분']=='총인구수 (명)', '구분'] = '합계'    # 총인구수 -> 합계
population.loc[population['구분']=='남자인구수 (명)', '구분'] = '남자'  # 남자인구수 -> 남자
population.loc[population['구분']=='여자인구수 (명)', '구분'] = '여자'  # 여자인구수 -> 여자

population.head()
```

칼럼 이름들이 깔끔하게 잘 정리되었네요~

### 7.3 인구 소멸 위기 지역 계산하고 데이터 정리하기  
인구 소멸 위기 지역을 알기 위해선 **65세 이상 노인 인구**와 **20-39세 여성 인구**를 알아야 한다고 했습니다. 지금의 데이터 구분 방식은 맞지 않으니 합할 건 합하고 추가할 건 추가해봅시다!

```python
population['20-39세'] = population['20 - 24세'] + population['25 - 29세'] + \
                        population['30 - 34세'] + population['35 - 39세']
    
population['65세이상'] = population['65 - 69세'] + population['70 - 74세'] + \
                        population['75 - 79세'] + population['80 - 84세'] + \
                        population['85 - 89세'] + population['90 - 94세'] + \
                        population['95 - 99세'] + population['100+']
            
population.head(10)
```

이제 필요한 데이터들은 얼추 만들어졌습니다. 데이터들을 정리하면 조금 더 보기 편할 것 같아요. piviot_table을 이용해 데이터들을 잘 정리해줍시다.

```python
pop = pd.pivot_table(population, 
                     index = ['광역시도', '시도'], 
                     columns = ['구분'],
                     values = ['인구수', '20-39세', '65세이상'])
pop.head()
```

이렇게 컬럼 2층, 인덱스 2개인 데이터프레임이 만들어졌습니다. 이제는 여기에다가 소멸비율을 계산하면 되겠죠! 20-39세 여성인구와 65세이상 합계를 사용해서 계산할 수 있습니다.

```python
pop['소멸비율'] = pop['20-39세','여자'] / (pop['65세이상','합계'] / 2)
pop.head()
```

코드 한 줄로 이렇게 많은 연산이 된다니... 신기하지 않나요?   
그럼 소멸 위기 지역이 어디인지를 새로운 칼럼을 만들어 넣어주겠습니다

```python
pop['소멸위기지역'] = pop['소멸비율'] < 1.0  # Boolean으로 표시해줄거에요(T or F)
pop.head()
```

그런데 이렇게 정리할 경우에 인덱싱하기가 굉장히 까다로워진다는 단점이 있습니다. 예를 들어 **강원도 강릉시의 20~39세 여성 인구**를 알고 싶으면 입력해야하는 조건 칼럼 조건 2개(20~39세, 여자), 로우에서 조건 2개(강원도, 강릉시)가 필요합니다. 이렇게 되면 후에 지도 시각화를 할 때 작성해야 하는 코드가 너무 복잡해져버리겠죠?

우선 reset_index()를 해주고요

```python
pop.reset_index(inplace=True) 
pop.head()
```

columns의 경우 **columns.get_level_values()** 라는 메서드를 쓸겁니다.

- [get_level_values()](https://www.it-swarm.dev/ko/python/%EB%8B%A4%EC%A4%91-%EC%9D%B8%EB%8D%B1%EC%8A%A4%EC%97%90%EC%84%9C-%EC%84%A0%ED%83%9D-pandas/1042031529/) (링크는 index.get_level_values()에 대한 내용이긴한데 동작 원리는 똑같습니다. 다만 동작이 index가 아닌 column에서 나타난다는 차이만 있어요)

```python
pop.columns
```

위를 보면 column이 두 겹으로 이루어진다는 것을 볼 수 있습니다. get_level_values()는 각각의 겹을 따로 떼어내 볼 수 있게 해주는 메서드라고 생각하면 됩니다. 예를 들어 columns.get_level_values(0)는 칼럼의 첫번째 겹을 보여달라는 말이에요

```python
pop.columns.get_level_values(0)
```

```python
pop.columns.get_level_values(1) # 칼럼의 2번째 겹을 보고 싶다면
```

get_level_values()를 통해 두 개로 구분되어 있던 칼럼을 하나로 합쳐주도록 하겠습니다!

```python
tmp_coloumns = [pop.columns.get_level_values(0)[n] + \
                pop.columns.get_level_values(1)[n] 
                for n in range(0,len(pop.columns.get_level_values(0)))]

pop.columns = tmp_coloumns

pop.head()
```

### 7.5 지도 시각화를 위해 지역별 고유 ID 만들기
장소를 시각하기 위해서는 어떻게 해야 할까요? json 파일의 내용과 위에 전처리해준 데이터를 하나의 데이터프레임 안에 넣어 사용하기에는 데이터 형식이 너무 다릅니다. 

대신 우리는 두 파일에 공통적인 ID를 만들고 그 ID를 매개로 하여 json의 데이터와 전처리해준 데이터를 하나로 엮어줄겁니다. 말이 조금 복잡한데요. 쉽게 말하자면 각각 '서울시 : pop = 100', '서울시 : lat = 127'라는 데이터가 있을 때 '서울시'를 ID로 만들어 'pop = 100 : lat = 127' 이라는 결과물을 만들어내는 것입니다.

json 파일에 있는 지역명을 ID로 설정하고 그에 맞게 데이터프레임을 수정해봅시다!

```python
pop['시도'].unique()
```

이게 무척 많은데, 보면 서울특별시 아래에 있는 동작구, 서대문구 등과 같은 광역시의 구가 있구요, 다른 한편으로는 안양시나 수원시에 있는 행정구가 있어요. 이것들을 다르게 표현해주면 더 명확하게 지도를 표시할 수 있을 것 같아요. 그래서 '광역시도'와 '시도'의 값을 우선적으로 합쳐서 '서울 강남'과 같이 표시하면 좋을 것 같습니다. 구이름이 두 글자면 '서울 중구'라고 표시하고, 일반 자치시의 경우에는 '충북 제천'과 같이 표기하면 좋을 것 같네요!

그런데 단순히 광역시도와 시구를 합친다고 하면,

```python
# 혜식님 코드: drawKorea()라는 함수를 만들어 준것! 
def drawKorea(targetData, blockedMap, cmapname):  # 여기 주의해서 보기!
    gamma = 0.75
    
    # 시각화 시에 소멸지역이 아닌 곳은 흰색으로 표시되게끔! 
    whitelabelmin = (max(blockedMap[targetData]) - 
                                     min(blockedMap[targetData]))*0.25 + \
                                                                min(blockedMap[targetData])

    datalabel = targetData

    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. 
        #(중구, 서구)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # 시도 경계 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
```

저희가 가지고 있는 여성 인구 데이터를 가지고 시각화를 해보려 합니다. 이게 자칫하면 가임기 여성 지도처럼 될 수 있어서 책 저자의 의도를 봤더니 '여성 인구가 너무 줄어 남녀 비율의 불균형 현상'을 보려했더라고요.

그 전에 앞서 만들어 놓은 drawKorea 함수를 살짝 손봐야해요. 후에 여성비를 만들 것인데, 이게 색상 팔레트 문제 때문에 기준을 0으로 하도록 해서 음수가 나올 수 있거든요. 그래서 tmp_max부분을 손봤어요. (근데 어차피 다 검은색으로 나오기 때문에 색은 중요하지 않지만, 데이터를 손 보면 음수가 뜰 수 있어서 오류가 나서 수정한 거에요)

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75

    whitelabelmin = 20.

    datalabel = targetData
    
    # 음수가 생기지 않도록! 
    tmp_max = max([ np.abs(min(blockedMap[targetData])), 
                                  np.abs(max(blockedMap[targetData]))])
    vmin, vmax = -tmp_max, tmp_max

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. 
        #(중구, 서구)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if np.abs(row[targetData]) > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # 시도 경계 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
```

이제 여성비를 계산하고 시각화를 해볼까요?

```python
pop['여성비'] = (pop['인구수여자']/pop['인구수합계'] - 0.5)*100
drawKorea('여성비', pop, 'RdBu')
```

0을 기준으로 계산을 했기 때문에 0 이상이면 여성비가 높고, 0이하면 여성비가 낮은 곳이라고 생각하면 됩니다. 웅진과 철원, 화천은 군부대 때문에 저런 극단적인 비율이 나온 것 같아요.

시계열로 봤으면 더욱 흥미로운 결과가 나왔을 것 같아요. 제 생각에는 전체적으로 빨간색이었다가 점점 빨간색이 옅어질 것 같아요. 핵심 변수는 가부장제(구체적으로 여아낙태)일 것 같고요.

이번에는 전체 20-30대 인구 중 20-30대 여성으로 데이터를 바꾸어서 시각화해보겠습니다.

```python
pop['2030여성비'] = (pop['20-39세여자']/pop['20-39세합계'] - 0.5)*100
drawKorea('2030여성비', pop, 'RdBu')
```

전체적으로 남성이 더 많네요. 서울과 성남, 용인, 세종과 같이 가족단위 가구가 많은 지역은 조금 덜 한 것 같습니다. 지방은 거의 20-30대는 남성이 더 많은 것 같습니다. 용진, 화천, 양구, 철원과 같이 군부대가 있는 지역의 경우 그 현상이 더욱 두드러지고요. 2015년에 20-30대면 1980-90년대 생 같은데, 그 때에는 아직도 남아 선호사상이 줄어들었다고 해도 꽤 작용한 것 같습니다.
